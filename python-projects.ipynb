```python
import tkinter as tk
root=tk.Tk()
def budget_game():   
    #WELCOME TO FIFA WORLD CUP
    print('Welcome to Fifa World Cup')

    #USER NAME
    User_Name=input('Enter your name: ')

    #PRINT USER NAME
    print('Hello!',User_Name)

    #USER BUDGET
    User_Budget=input('Enter your budget for buying ticket: ')

    #IF USER BUDGET IS MORE THAN 25000 -- HURRAY! YOU CAN WATCH THE MATCH WITH VIP SEATS
    if(int(User_Budget)>=25000):
        print('Hurray! You can watch the match with VIP seats')

    #IF USER BUDGET IS IN BETWEEN 10000 to 25000 -- HURRAY! YOU CAN WATCH THE MATCH WITH SEMI-VIP SEATS
    elif(int(User_Budget)>=10000 and int(User_Budget)<25000):
        print('Hurray! You can watch the match with Semi-VIP seats')

    #IF USER BUDGET IS IN BETWEEN 1000 to 10000 -- HURRAY! YOU CAN WATCH THE MATCH
    elif(int(User_Budget)>=1000 and int(User_Budget)<10000):
        print('Hurray! You can watch the match')

    #ELSE SORRY! YOU CANT WATCH THE MATCH
    else:
        print('Sorry! You cant watch the match')

def heads_and_tails():    
    #WELCOME TO H&T GAME
    print('Welcome to Coin Game')

    #USER INPUT
    user_input=input('Choose Heads or Tails: ')

    #PRINT USER INPUT
    print('User Input: ',user_input)
    
    #IF USER INPUT IS HEADS OR TAILS 
    if user_input.lower()=='heads'or'tails':
        #THROW COIN
        import random
        coin_result=random.choice(['heads','tails'])
        print('Coin result:',coin_result)

        #IF COIN RESULT == USER INPUT - YOU WON
        if(coin_result==user_input):
            print('Hurray! you won')

        #ELSE - YOU LOOSE
        else:
            print('You Loose')
    #ELSE - INVALID INPUT
    else:
        print('Invalid Input! Please enter again')
        heads_and_tails()

def dice_game():
    #WELCOME TO DICE GAME
    print('Welcome to Dice Game')

    #USER NAME
    user_name=input('Enter your name: ')

    #PRINT USER NAME
    print('Hello!',user_name)

    #USER INPUT
    user_input=input('Choose a number in between 1 to 6: ')

    #THROW DICE
    import random
    dice_result=random.randrange(1,7)

    #DICE RESULT
    print('Dice_Result: ',dice_result)

    #IF USER INPUT IS EQUAL TO DICE RESULT - YOU WIN
    if(user_input==dice_result):
        print('You Win')

    #ELSE - YOU LOOSE
    else:
        print('You Loose')
        
def cube_number():
    #CUBE OF GIVEN NUMBER
    print('Cube of Given Number')

    #USER INPUT
    user_input=int(input('Enter a number: '))

    #USING FOR LOOP
    for x in range(user_input+1):

    #FIND CUBE
        cube=x**3

    #PRINT THE CUBE NUMBER
        print('The cube of',x,'is',cube)
        
def square_number():
    #SQUARE OF GIVEN NUMBER
    print('Square of Given Number')

    #USER INPUT
    user_input=int(input('Enter a number: '))

    #USING FOR LOOP
    for x in range(user_input+1):

    #FIND SQUARE
        square=x**2

    #PRINT THE SQUARE NUMBER
        print('The square of',x,'is',square)

def factorial():
    #USER INPUT
    user_input=int(input('Enter a number:'))
    fact=1

    #USING FOR LOOP
    for x in range(1,user_input+1):

    #FIND FACTORIAL
        fact=fact*x

    #PRINT THE FACTORIAL NUMBER
    print('The factorial of',x,'is',fact)
    
def fibonacci_series():
    #WELCOME TO FIBONACCI SERIES
    print('Welcome to Fibonacci Series')

    #USER INPUT
    user_input=int(input('Enter a number: '))
    A=0
    B=1

    #USING FOR LOOP
    for x in range(1,user_input+1):

    #USING IF STATEMENT
        if(x<=1):
            FS=x

    #USING ELSE STATEMENT
        else:
            FS=A+B
            A=B
            B=FS

        if(FS<user_input):

    #PRINT THE FIBONACCI SERIES
            print(FS)
        else:
            break
            
def password_generator():
    #IMPORT RANDOM AND STRING
    import random
    import string

    #CREATE A PASSWORD GENERATOR
    print('Create a Password Generator')

    #USER INPUT, CHARACTER LIST AND PASSWORD
    user_input=input('Enter Password Length: ')
    character_list=string.ascii_letters+string.punctuation+string.digits
    password=''

    #USING FOR LOOP
    for x in range(int(user_input)):
        random_char=random.choice(character_list)
        password=password+(random_char)

    #PRINT THE RANDOM PASSWORD
    print('The random password is',(password))
    
    #GENERATE AGAIN
    generate_again=input('Do you want to generate again? ')
    
    #IF GENERATE AGAIN == YES
    if generate_again.lower()=='yes':
        password_generator()
    #ELSE -- OK NEVERMIND
    else:
        print('Ok Nevermind!')
    
def countdown_timer():
    #WELCOME TO COUNTDOWN TIMER
    print('Welcome to Countdown Timer')

    #IMPORT TIME
    import time
    my_time=int(input('Enter the time in seconds: '))

    #USING FOR LOOP
    for x in range(my_time,0,-1):
        seconds=x%60
        minutes=int(x/60)%60
        print(f'00:{minutes:02}:{seconds:02}',end='\n')
        time.sleep(1)

    #PRINT TIMES UP FOR COUNTDOWN
    print('Times Up!')
    
def mind_game(): 
    #WELCOME TO MIND GAME
    print('Hello! I am an AI and I will read your mind')
    #USER INPUT
    user_input=input('Do you want to play a game: ')
    #IMPORT TIME
    import time
    #IF USER INPUT IS EQUAL TO YES -- THEN PRINT AND START THE GAME
    if user_input.lower()=='yes':
        #GUESS A NUMBER FROM 1-500
        print('Ok guess a number in your mind from 1-500')
        time.sleep(5)
        #MULTIPLY THE GUESSED NUMBER BY 2
        print('Multiply the guessed number by 2')
        time.sleep(5)
        #THEN ADD THE GUESSED NUMBER WITH 50
        print('Then add the guessed number with 50')
        time.sleep(5)
        #NOW DIVIDE THE GUESSED NUMBER BY 2
        print('Now divide the guessed number by 2')
        time.sleep(5)
        #MINUS THE NUMBER WITH THE ORIGINAL GUESSED NUMBER
        print('Minus the number with the original guessed number')
        time.sleep(5)
        #COMPUTER STARTED TO GUESS
        print('Umm...')
        time.sleep(2)
        #COMPUTER IS GUESSING
        print('Ok, I am guessing...')
        time.sleep(2)
        #THE ANSWER IS PROCESSING
        print('The number might be...')
        time.sleep(2)
        #COMPUTER DISPLAYED THE ANSWER
        print('Yes, the answer is 25!!')
        time.sleep(2)
        mind_game()
    #ELIF USER INPUT IS EQUAL TO NO -- THEN DONT START THE GAME
    elif user_input.lower()=='no':
        #OK NEVERMIND DONT START THE GAME
        print('Ok Nevermind!!')
    #ELSE USER INPUT IS EQUAL TO SOMETHING ELSE
    else:
        #INVALID INPUT PLEASE ENTER YES OR NO
        print('Invalid!! Input please enter yes or no')
        
def coffee_shop():
    menu={'Espresso':350,
        'Cappucino':400,
        'Latte':450,
        'Mocha':500,
        'Flat White':250}
    payment_list=[]
    addition_values=[]

    def display_menu():
        '''DISPLAYING THE MENU TO THE USER'''
        print('Welcome to Starbucks!')
        print('Menu:')
        print('*'*20)

        for i,j in menu.items():
            print(i,j)
        print('*'*20)
    def choose_drink():
        '''USER CHOOSING DRINK BY THE MENU'''
        global user_choice
        user_choice=input('Enter your drink choice: ')
        print('You have ordered a',user_choice,'!')
        if user_choice.title() in menu.keys():
            payment_list.append(user_choice.title())
        else:
            pass
        return user_choice
    def verify_order():
        '''VERIFYING THE ORDER THAT WAS GIVEN BY THE USER'''
        if user_choice.title() in menu.keys():
            print('Thank you for ordering')
            order_proceed='yes'
        else:
            print('Invalid item! Please order again')
            order_proceed='no'
        return order_proceed
    def preparing_drink():
        '''PREPARING DRINK FOR THE USER'''
        import time
        print('Preparing your',user_choice,'...')
        time.sleep(2)
        print('Done! Enjoy your drink!')

    def order_again():
        '''ASKING THE USER TO ORDER AGAIN'''
        user_input=input('Would you like to order something else ? ')
        if user_input.lower()=='yes':
            display_menu()
            choose_drink()
            order_again()
        elif user_input.lower()=='no':
            print('Ok nevermind')
    def payment():
        '''DO PAYMENT FOR THE ORDER'''
        global user_payment
        for i in payment_list:
            addition_values.append(menu.get(i))
        total_order=' , '.join(payment_list) 
        print(f'Your {total_order} costs {sum(addition_values)}')
        user_payment=int(input('Enter payment amount: '))
    def change():
        '''GIVE CHANGE TO THE USER IF USER PAYMENT IS MORE'''
        user_change=abs(sum(addition_values)-user_payment)
        print('Thank you for your payment! Your change is',user_change)
    def pending():
        '''GIVE MORE MONEY IF USER PAYMENT IS LESS'''
        pending_money=abs(user_payment-sum(addition_values))
        print('Your payment is less of:',pending_money)
    def ending():
        '''ENDING THE COFFEE SHOP PROCESS'''
        print('Thank you for visiting Starbucks!')                      

    def full_coffee_func():
        '''PRINTING THE EARLIER FUNCTION HERE TO GET THE OUTPUT'''
        display_menu()
        choose_drink()
        continue_run=verify_order()

        if continue_run=='yes':
            preparing_drink()
            order_again()
            payment()
            if sum(addition_values)>user_payment:
                pending()
            elif sum(addition_values)==user_payment:
                pass
            else:
                change()
            ending()
        else:
            full_coffee_func()

    full_coffee_func()
    
root.title('Python Projects')
root.geometry('500x400')
label=tk.Label(root, text='PYTHON PROJECTS',font=('Times New Roman',20),)
label.place(x=120,y=30)
label=tk.Label(root, text='Made By: Hrushikesh')
label.place(x=170,y=70)

button=tk.Button(root,text='Budget Game', command=budget_game)
button.place(x=75,y=120)

button=tk.Button(root,text='H&T Game',command=heads_and_tails)
button.place(x=200,y=120)

button=tk.Button(root,text='Dice Game',command=dice_game)
button.place(x=330,y=120)

button=tk.Button(root,text='Cube Number',command=cube_number)
button.place(x=75,y=170)

button=tk.Button(root,text='Square Number',command=square_number)
button.place(x=190,y=170)

button=tk.Button(root,text='Factorial',command=factorial, width=10)
button.place(x=330,y=170)

button=tk.Button(root,text='Fibonacci Series',command=fibonacci_series)
button.place(x=70,y=220)

button=tk.Button(root,text='Password Generator',command=password_generator)
button.place(x=180,y=220)

button=tk.Button(root,text='Countdown Timer',command=countdown_timer)
button.place(x=310,y=220)

button=tk.Button(root,text='Mind Game',command=mind_game)
button.place(x=80,y=270)

button=tk.Button(root,text='Coffee Shop',command=coffee_shop)
button.place(x=330,y=270)

root.mainloop()
```

## Threading to do different tasks

```python
#IMPORT THREADING AND TIME
!pip3 install thread6
import threading
import time

def play_music():
    for _ in range(3):
        #PRINT PLAYING MUSIC
        print('Playing Music')
        time.sleep(1) #SIMULATE SOME TIME PASSING
        
def color_picture():
    for _ in range(3):
        #PRINT COLOURING PICTURE
        print('Colouring Picture')
        time.sleep(3) #SIMULATE SOME TIME PASSING
        
def eat_food():
    for _ in range(6):
        #PRINT EATING FOOD
        print('Eating Food')
        time.sleep(0.5) #SIMULATE SOME TIME PASSING
        
#CREATING THREADS FOR THE TASK       
music_thread = threading.Thread(target=play_music)
colour_thread = threading.Thread(target=color_picture)
food_thread = threading.Thread(target=eat_food)

#STARTING THE THREADS
music_thread.start
colour_thread.start
food_thread.start

#WAITING FOR THREADS TO FINISH
music_thread.join
colour_thread.join
food_thread.join

play_music()
color_picture()
eat_food()
#PRINT ALL TASKS ARE DONE
print('All tasks are done')
```

## Duck Game

```python
class Duck:
    def __init__(self, name):
        self.name = name
        #DUCK START WITH 100 ENERGY
        self.energy = 100
        
    def __str__(self):
        return f"{self.name} ğŸ¦† (Energy: {self.energy})"
    
    def quack(self):
        '''MAKING THE DUCK QUACK WHILE DECREASING OR INCREASING ITS ENERGY BY 5'''
        if self.energy >= 5:
            print(f"{self.name} says Quack! ğŸ¦†")
            self.energy -= 5
        else:
            print(f"{self.name} is too tired to quack. ğŸ˜«")
    
    def eat(self):
        '''WHEN THE DUCK EATS THE ENERGY WILL BE ADDED BY 10'''
        print(f"{self.name} is eating. ğŸ” Yum!")
        self.energy += 10
        
    def sleep(self):
        '''WHEN THE DUCK SLEEPS THE ENERGY WILL BE ADDED BY 15'''
        print(f"{self.name} is sleeping. ğŸ˜´ Zzz...")
        self.energy +=15
        
    def play(self):
        '''WHEN THE DUCK PLAYS THE ENERGY WIL BE REDUCED BY 10'''
        if self.energy >= 10:
            print(f"{self.name} is playing. ğŸ¥µ Whee!")
            self.energy -= 10
        else:
            print(f"{self.name} is too tired to play. ğŸ˜«")
            
def main():
    #CREATE A DUCK NAMED "DUCKY"
    ducky = Duck("Donald Duck")
    
    #USE DUNDER METHOD __str__ TO DISPLAY THE DUCK'S STATUS
    print(ducky) #OUTPUT: "DUCKY ğŸ¦†(ENERGY: 100)"
    
    #INTERACT ACTIVITIES WITH THE DUCK
    ducky.quack()  #OUTPUT: "DUCKY SAYS QUACK! ğŸ¦†"
    ducky.eat()    #OUTPUT: "DUCKY IS EATING. ğŸ” Yum!"
    ducky.sleep()  #OUTPUT: "DUCKY IS SLEEPING. ğŸ˜´Zzz..."
    for x in range(5):
        ducky.play()  #OUTPUT: "DUCK IS PLAYING. ğŸ¥µ Whee!"
        
    #DISPLAY DUCK'S STATUS AGAIN
    print(ducky) #OUTPUT: "DUCKY ğŸ¦† (ENERGY: 95)"
    
if __name__ == "__main__":
    main()
```

## Created a contact list

```python
#PRINT WELCOME TO THE WORLD OF CONTACTS
print('Welcome to the world of contacts!')
contacts={}
#FUNCTION TO ADD A NEW CONTACT
def add_contacts():
    phone=int(input('Enter contact phone number: '))
    name=input('Enter contact name: ')
    contacts[name]=phone
    print(f'{name} has been added to your contacts!')
#FUNCTION TO ADD A NEW CONTACT AGAIN
def add_contacts_again():
    add_again=input('Would you like to add more contacts? ')
    if add_again.lower()=='yes':
        add_contacts()
        add_contacts_again()
        search_contacts()
        display_contacts()
    else:
        print('Ok Nevermind!')
#FUNCTION TO SEARCH FOR A CONTACT
def search_contacts():
    name=input('Enter a contact name to search: ')
    if name in contacts:
        print(f'Phone number for {name}: {contacts[name]}')
    else:
        print(f'{name} not found in contacts.')
#FUNCTION TO DISPLAY CONTACTS
def display_contacts():
    if contacts:
        print('Your contacts list: ')
        for name, phone in contacts.items():
            print(f'{name}: {phone}')
    else:
        print('Your contacts list is empty!')
add_contacts()
add_contacts_again()
search_contacts()
display_contacts()
```

## To-Do List

```python
#IMPORTING LIBRARY
import tkinter as tk
root=tk.Tk()

#ADD THE TASK FOR THE USER
def add_task():
    user_task=entry.get()
    listbox.insert(tk.END, user_task)
    
#MARK COLOR FORM DISPLAY TASK
def mark_complete():
    click_task=listbox.curselection()
    click_task=click_task[0]
    listbox.itemconfig(click_task, {'bg': 'cyan'})
    
#REMOVE THE TASK FROM ADD TASK
def remove_complete():
    click_task=listbox.curselection()
    listbox.delete(click_task)
    
#NAMING THE TITLE AND MEASURING ITS SIZE
root.title('To-Do List')
root.geometry('250x350')

#ENTRY WIDGET FOR DISPLAYING THE RESULT
entry=tk.Entry(root)
entry.pack()

#ADDING BUTTON TO INSERT THE VALUES
button=tk.Button(root,text='Add Task', command=add_task)
button.pack()
button.place(x=100,y=40)

#CREATING AND PLACING THE LISTBOX TO DISPLAY THE OUTPUT
listbox=tk.Listbox(root)
listbox.pack()
listbox.place(x=65,y=80)

#ADDING BUTTON TO GET THE TASK VALUE MARKED
button=tk.Button(root,text='Mark Complete', command=mark_complete)
button.pack()
button.place(x=85,y=250)

#ADDING BUTTON TO GET THE TASK VALUE REMOVED
button=tk.Button(root,text='Remove Completed', command=remove_complete)
button.pack()
button.place(x=75,y=290)
root.mainloop()
```
